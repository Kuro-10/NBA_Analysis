```python
r = requests.get(url=test_url).json()
```

```python
del temp_df1, temp_df2, temp_df3
```

```python
df_cols = ['Year', 'SeasonType'] + table_header
```

```python
df = pd.DataFrame(columns=df_cols)
seasons_type = ['Regular%20Season', 'Playoffs']
years = ['2012-13','2013-14','2014-15','2015-16','2016-17','2017-18' ,'2018-19','2019-20', '2020-21', '2021-22', '2022-23', '2023-24']

begin_loop = time.time()

for y in years:
    for s in seasons_type:
        test_url = f"https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season={y}&SeasonType={s}&StatCategory=PTS"
        r = requests.get(url=test_url).json()   
        temp_df1 = pd.DataFrame(r['resultSet']['rowSet'], columns=table_header)
        temp_df2 = pd.DataFrame({'Year': [y for i in range(len(temp_df1))],
                                 'SeasonType': [s for i in range(len(temp_df1))]})
        temp_df3 = pd.concat([temp_df1, temp_df2], axis=1)
        df = pd.concat([df, temp_df3], axis=0)
        print(f'Finished {y} {s}')
        lag = np.random.uniform(low=5, high=40)
        print(f'... waiting {round(lag,1)} seconds')
        time.sleep(lag)
        
print(f'Finished in {round((time.time()-begin_loop)/60,2)} seconds')
df.to_excel('NBA_Stats.xlsx', index=False)
```
